```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model = load_model('model.h5')

with open('label_encoder_gender.pkl' , 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('onehot_Geography.pkl' , 'rb') as file:
    onehot_Geography = pickle.load(file)

with open('standard_scalar.pkl' , 'rb') as file:
    standard_scalar = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
geo = onehot_Geography.transform(input_df[['Geography']]).toarray()
geo_df = pd.DataFrame(geo,columns=onehot_Geography.get_feature_names_out(['Geography']))
```

```python
input_df = pd.concat([input_df.drop("Geography" , axis=1) , geo_df] , axis=1)
```

```python
input_scaled=standard_scalar.transform(input_df)
```

```python
prediction=model.predict(input_scaled)
prediction_proba = prediction[0][0]
print(prediction_proba)
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
